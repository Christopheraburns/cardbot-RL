```python
import gym
import sys
import numpy as np
import random
```

States Tuple:

0 = Players current sum: [0,31] i.e. 32 states
    0 = no cards
    1-11 for first card
    1-11 for second card
    
0 = Dealer's face up card: [1,10] i.e. 10 states
0 = Whether the player has a usable ace or not: [0,1] i.e. 2 states


Initial state = 2 player cards

Actions:

Either stay or hit: [0,1] i.e 0 for stay , 1 for hit

```python
env = gym.make('Blackjack-v0')
print(env.observation_space)
print(env.action_space)
```

```python
initial_state = env.reset()
print(initial_state)
print("player 2 card sum: {} ({} Ace), Dealer face-up card: {}".format(initial_state[0], 
                                                                "Yes" if initial_state[2] is True else "No", 
                                                                initial_state[1]))
#action = input("Choose an action.  Enter 0 for stay, 1 for hit")
```

```python
# Look at a random policy
win = 0
loss = 0

for round in range(200):  # play 1000 rounds at random guessing (unrealistic, but helps build a baseline)
    state = env.reset()
    while True:
        #print(state)
        action = env.action_space.sample() # Random action
        state, reward, done, info = env.step(action)
        if done:
            print(state)
            if reward > 0:
                win = win + 1
            else:
                loss = loss + 1
            break

print("{} wins vs {} losses with random policy".format(win, loss))
```
